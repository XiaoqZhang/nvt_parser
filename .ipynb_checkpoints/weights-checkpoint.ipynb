## Get supercell

```python
def get_supercell(structure):
    with open("nvt_results/%s/simulation.input" %structure, "r") as f_input:
        for line in f_input:
            if re.search("UnitCells", line):
                s1, s2, s3 = line.split()[1], line.split()[2], line.split()[3]
    supercell = [int(s1), int(s2), int(s3)]
    return supercell
```

### Old way

```python
# read ff from .def
def read_ff():
    with open("parameters/UFFTraPPE/force_field_mixing_rules.def") as file:
        ff = file.readlines()
        ff = pd.DataFrame([line.split() for line in ff], columns=['element', 'lj', 'epsilon', 'sigma'])
        ff = ff.drop(columns=['lj'])
        col = ["sigma", "epsilon"]
        ff[col] = ff[col].apply(pd.to_numeric)
        ff["element"] = [re.sub("\_$", "", i) for i in list(ff["element"])]
    return ff

uff = read_ff()
```

```python
def lb_rule(df_ff, gas_sigma, gas_epsilon):
    df_ff["lb_sigma"] = [math.sqrt(value * gas_sigma) for value in list(df_ff["sigma"])]
    df_ff["lb_epsilon"] = [0.5 * (value + gas_epsilon) for value in list(df_ff["epsilon"])]

    return df_ff

uff = lb_rule(uff, ch4_sigma, ch4_epsilon)
```

### Old method

```python
def sites(structure, supercell, df_ff):
    
    # Read framework lattice
    crys_lattice = Structure.from_file("nvt_results/%s/Movies/System_0/Framework_0_final.vasp" %structure)
    framework_sites = crys_lattice.sites

    # gas sites
    with open("nvt_results/%s/Movies/System_0/Movie_%s_%d.%d.%d_298.000000_0.000000_allcomponents.pdb" %(structure, structure, supercell[0], supercell[1], supercell[2]), "r") as f_molecule:
        f_molecule_lines = f_molecule.readlines()

        model_count = 0
        molecule_lines = []       
        
        for line in f_molecule_lines:
            if "MODEL" in line:
                model_count += 1
            if "ATOM" in line:
                molecule_lines.append(line)
        df_molecule = pd.DataFrame([line.split() for line in molecule_lines], 
                                    columns=['ATOM', 'No', 'Element', 'MOL', 'x', 'y', 'z', 'Occ', 'Ueq', 'Element_label'])
        df_molecule.drop(columns=['ATOM', 'MOL'], inplace=True)
        col = ['No', 'x', 'y', 'z', 'Occ', 'Ueq'] 
        df_molecule[col] = df_molecule[col].apply(pd.to_numeric)

    molecule_sites = []
    for x, y, z in zip(df_molecule.x, df_molecule.y, df_molecule.z):
        molecule_sites.append(PeriodicSite(species={"CH4": 0}, coords=np.array([x, y, z]), 
                                lattice=crys_lattice.lattice, coords_are_cartesian=True))
    
    # initialize framwork site properties
    for site in framework_sites:
       site.properties = {"count": np.zeros(model_count), "distance": np.zeros(model_count), "lj": np.zeros(model_count), 
                            "weight": 0, 
                            "lb_sigma": df_ff[df_ff["element"] == site.specie.symbol]["lb_sigma"].values, 
                            "lb_epsilon": df_ff[df_ff["element"] == site.specie.symbol]["lb_epsilon"].values}

    return crys_lattice, molecule_sites
```

```python
crys_lattice, mol_sites = sites(structure, sc, uff)
```

### Read molecule movies and get the fractional coordinates in the unit cell

```python
with open("nvt_results/%s/Movies/System_0/Movie_%s_%d.%d.%d_298.000000_0.000000_allcomponents.pdb" %(structure, structure, sc[0], sc[1], sc[2])) as file:
    data = file.readlines()
    atom_pos = np.array([np.array([float(i) for i in line.split()[4:7]]) for line in data if "ATOM" in line])
```

```python
# Get atoms in threshold sphere
threshold = 4.0
sphere_sites = np.array([un_cry.get_sites_in_sphere(pt=pos, r=threshold) for pos in atom_pos], dtype=object)
```

```python
# Calculate distance between gas and surrounding framework atoms
dist = np.array([np.array([site.distance_from_point(atom_pos[idx]) for site in sphere_sites[idx]]) 
                    for idx in range(len(atom_pos))], dtype=object)
```

```python
# Calculate LJ potential

def lj(sigma_lb, epsilon_lb, distance):
    #distance_reduced = distance/sigma_lb
    u_lj = 4 * epsilon_lb * ((sigma_lb/distance)**12 - (sigma_lb/distance)**6)
    return u_lj

potential = np.array([np.array([lj(sphere_sites[i][j].properties["lb_sigma"], sphere_sites[i][j].properties["lb_epsilon"], dist[i][j]) 
                for j in range(len(sphere_sites[i]))]).flatten() for i in range(len(atom_pos))], dtype=object)
```

```python
for idx in range(len(atom_pos)):
    if len(sphere_sites[idx]) != len(dist[idx]):
        print("error in distance")
    if len(sphere_sites[idx]) != len(potential[idx]):
        print("error in potential")
```

```python
def lj(sigma_lb, epsilon_lb, distance):
    #distance_reduced = distance/sigma_lb
    u_lj = 4 * epsilon_lb * ((sigma_lb/distance)**12 - (sigma_lb/distance)**6)
    return u_lj
```

```python
molecule_coords = [mol.coords for mol in mol_sites]
```

```python
norm_lj = []

for idx in range(len(molecule_coords)):
    
    # calculate the lj potential
        for atom_idx in range(len(crys_lattice.sites)):
            if crys_lattice.sites[atom_idx].specie.symbol != "H":
                dist = crys_lattice.sites[atom_idx].distance(mol_sites[idx])
                sigma_lb, epsilon_lb = crys_lattice.sites[atom_idx].properties["lb_sigma"], crys_lattice.sites[atom_idx].properties["lb_epsilon"]
                crys_lattice.sites[atom_idx].properties["lj"][idx] = lj(sigma_lb, epsilon_lb, dist)
    
    # count the gas molecule within the sphere
        site_list = crys_lattice.get_sites_in_sphere(pt=molecule_coords[idx], r=3.8)
        for i in site_list:
            if i.specie.symbol != "H":
                i.properties["count"][idx] = 1
                norm_lj.append(i.properties["lj"][idx])
```

## Calculate the weights for each atom

```python
all_lj = np.array([(site.properties["lj"]) for site in crys_lattice.sites])
all_count = np.array([(site.properties["count"]) for site in crys_lattice.sites])
```

```python
all_lj_norm = np.zeros([all_lj.shape[0], all_lj.shape[1]])

min = np.array(norm_lj).min()
max = np.array(norm_lj).max()

for i in range(all_lj.shape[0]):
    for j in range(all_lj.shape[1]):
        all_lj_norm[i,j] = 1- 2 * (all_lj[i,j]-min) / (max-min)
```

```python
def read_framework(structure, supercell):
    with open("%s/%s.cif" %(structure,structure)) as f_framework:
        f_framework_lines = f_framework.readlines()
        f_framework_lines = f_framework_lines[23:]
        df_framework = pd.DataFrame([line.split() for line in f_framework_lines], 
                                    columns=['atom', 'label', 'x', 'y', 'z', 'charge'])
        df_framework.drop(columns=['x', 'y', 'z', 'charge'], inplace=True)       
        total_atom_num = len(df_framework)
        uc_atom_num = int(total_atom_num/(supercell[0]*supercell[1]*supercell[2]))
        
        # label the atoms in the framework here
        element = df_framework['atom'].unique().tolist()
        element_count = [0] * len(element)
        for atom_idx in range(len(df_framework)):
            for ele_idx in range(len(element)):
                if df_framework['atom'][atom_idx] == element[ele_idx]:
                    element_count[ele_idx] += 1
                    df_framework.loc[atom_idx, ('label')] = df_framework.loc[atom_idx, ('label')] + str(element_count[ele_idx])

    return df_framework
'''
        for uc_idx in range(supercell[0]*supercell[1]*supercell[2]):
            if uc_idx > 0:
                for atom_idx in range(uc_atom_num):
                    df_framework.loc[atom_idx + uc_atom_num * uc_idx, ('label')] = df_framework.loc[atom_idx, ('label')]
'''
    
```
